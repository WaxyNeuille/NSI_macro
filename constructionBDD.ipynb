# Table Consommation

```python
import sqlite3

bdd = sqlite3.connect("DB.db")
curseur = bdd.cursor()

requete = """
DROP TABLE IF EXISTS Consommation;
"""
curseur.execute(requete)

requete = """
CREATE TABLE Consommation (
    Date DATE,
    poid_Aliment FLOAT,
    IDutilisateur INT,
    IDaliment INT,
    PRIMARY KEY (Date, poid_Aliment, IDutilisateur, IDaliment),
    FOREIGN KEY (IDutilisateur) REFERENCES Utilisateurs(id) ON DELETE CASCADE,
    FOREIGN KEY (IDaliment) REFERENCES Aliments(IDAliment) ON DELETE CASCADE
);
"""
curseur.execute(requete)

bdd.commit()
bdd.close()
print("La relation a bien été créée.")
```
